```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_json('data-viz-challenge/data.json')
data.head()
```

```python
data.iloc[0][0]
```

```python
Category = []
Event_name = []
Gender = []
Age = []
Marital_Status = []
Client_time = []
City = []
State = []
Long = []
Lat = []
Device = []
Amount = []
```

```python
for i in range(len(data)):
    Category.append(data.iloc[i][0]['category'])
    Gender.append(data.iloc[i][0]['gender'])
    Age.append(data.iloc[i][0]['age'])
    Marital_Status.append(data.iloc[i][0]['marital_status'])
    Client_time.append(data.iloc[i][0]['client_time'])
    City.append(data.iloc[i][0]['location']['city'])
    State.append(data.iloc[i][0]['location']['state'])
    Long.append(data.iloc[i][0]['location']['longitude'])
    Lat.append(data.iloc[i][0]['location']['latitude'])
    Device.append(data.iloc[i][0]['device'])
    if data.iloc[i][0]['event_name'] == 'Fund Project':
        Amount.append(data.iloc[i][0]['amount'])
        Event_name.append(1)
    else:
        Amount.append(0)
        Event_name.append(0)
```

```python
d = pd.DataFrame()
Time = []
Tim = pd.to_datetime(Client_time , unit = 's')
for t in Tim:
    Time.append(str(t)[:10])
d['Category'] = Category
d['Fund'] = Event_name
d['Amount'] = Amount
d['Gender'] = Gender
d['Age_Range'] = Age
d['Marital_Status'] = Marital_Status
d['Client_time'] = Time
d['City'] = City
d['State'] = State
d['Long'] = Long
d['Lat'] = Lat
d['Device'] = Device
d.head()
```

## Checking the Sport and Environment category

```python
import seaborn as sns
```

```python
df= d[(d['Category']=='Sports') | (d['Category']=='Environment')]
```

```python
fig, (fig1,fig2) = plt.subplots(1,2,figsize=(10,5))
sns.barplot(data=df, x='Age_Range',y='Fund', ax=fig1)
sns.countplot(data=df, x='Age_Range',hue='Fund', ax=fig2)
```

From the count plot we infer that
maximum people fund in the age group of 18-24 
Then in descending order 35-44 / 45-44 / 25-34 / 55+ age groups fund.

From the bar plot we see that out of all the 18-24 age group people the chances of few of them funding is high. 

18-24 is the best bet as the number of people in it is also > than all the other age groups

```python
fig, (fig1,fig2) = plt.subplots(1,2,figsize=(10,5))
sns.barplot(data=df, x='Gender',y='Fund', ax=fig1)
sns.countplot(data=df, x='Gender',hue='Fund', ax=fig2)
```

Number of females funding is slightly greater than the number of males funding
And very few of the other gender fund.

Alhough the Unidentified Gender's probablity of funding is much higher compared to others their number is quite low.

Therefore it would be safe to say that there is a high chance that females will fund,

```python
fig, (fig1,fig2) = plt.subplots(1,2,figsize=(10,5))
sns.barplot(data=df, x='Marital_Status',y='Fund', ax=fig1)
sns.countplot(data=df, x='Marital_Status',hue='Fund', ax=fig2)
```

We can see from the countplot that the number of married people viewing/funding projects is much higher than the number of single people.

The bar plot also suggests that married people have a higher probablity of funding

Hence Married people will also be our target

```python
fund_loc_lon = []
fund_loc_lat = []
view_loc_lon = []
view_loc_lat = []

fund_time = []
view_time = []

for i in range(len(df)):
    if df.iloc[i]['Fund'] == 0:
        view_loc_lon.append(df.iloc[i]['Long'])
        view_loc_lat.append(df.iloc[i]['Lat'])
        view_time.append(df.iloc[i]['Client_time'])
    else:
        fund_loc_lon.append(df.iloc[i]['Long'])
        fund_loc_lat.append(df.iloc[i]['Lat'])
        fund_time.append(df.iloc[i]['Client_time'])
```

```python
sns.scatterplot(x=view_loc_lon , y=view_loc_lat, color='r')
```

```python
sns.scatterplot(x=fund_loc_lon , y=fund_loc_lat , color='b')
```

```python
sns.scatterplot(x=view_loc_lon , y=view_loc_lat, color='r')
sns.scatterplot(x=fund_loc_lon , y=fund_loc_lat , color='b')
```

When we plot on a scatter graph, the longitude and latitude cordinates for people who viewed (red) and people who fund (blue) ,  we notice that there is no major difference between them. It appears as if both are the same cordinates considering the similarity in the two.

Hence I conclude that the location of the person doesn't matter in deciding wether the person will fund or view.
